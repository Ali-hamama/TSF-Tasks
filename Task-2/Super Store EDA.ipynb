*we can see that there are duplicates in data that should be drop out to avoid bias in the analysis process*

```python
# drop duplicates from data
data_1 = raw_data.drop_duplicates(keep='first',ignore_index=True)
data_1.shape
```

```python
# Bar chart for state's profit

state_profit = data_1.groupby("State")[['Profit']].sum()

color = []
for idx,profit in state_profit.iterrows():
    if(profit[0]<=0):
        color.append("red")
    elif(profit[0]>0 and profit[0]< 10000):
        color.append("orange")
    else:
        color.append("green")

state_profit.plot.bar(y="Profit",figsize=(20,5),color=color)

red_patch = mpatches.Patch(color='red',label="<= 0")
orange_patch = mpatches.Patch(color='orange',label="> 0 and < 10000")
green_patch = mpatches.Patch(color='green',label="> 10000")

plt.title("Profit By State")
plt.ylabel("Profit")
plt.yticks(range(-20000,80001,20000),['-20k','0','20k','40k','60k','80k'])
plt.legend(handles=[red_patch,orange_patch,green_patch])
plt.show()
```

#### We can see in the previous chart there are about 10 states do NOT make any profit and instead they incur LOSSES.
* we'll dive deeper to see what those countries have in common.

```python
state_group = data_1.groupby(["State","Region"]).sum().reset_index()
losses_countries = state_group[state_group["Profit"] <= 0]
losses_countries
```
